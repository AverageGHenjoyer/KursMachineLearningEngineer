```python
import pandas as pd
```

```python
pd.read_csv("./data/winequality-red.csv", sep=";")
```

```python
pd.read_csv("./data/winequality-white.csv", sep=";")
```

```python
wine_quality_df = pd.concat([
    pd.read_csv("./data/winequality-white.csv", delimiter=";").assign(color="white"),
    pd.read_csv("./data/winequality-red.csv", delimiter=";").assign(color="red"),
    
])
wine_quality_df.sample(5)
```

```python
wine_quality_df.loc[wine_quality_df['color'] == "red"].count()
```

```python
wine_quality_df.corr() \
    .style.background_gradient(cmap="coolwarm")
```

```python
wine_quality_df.describe().T
```

```python
wine_quality_df.plot.box(figsize=(25, 7.5),rot=90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numeric_columns_df = wine_quality_df.select_dtypes(include=np.number)
```

```python
fig, ax = plt.subplots(6, 2, figsize=(16.0, 32.0))
flat_ax = ax.flatten()
for idx, column_name in enumerate(numeric_columns_df.columns):
    wine_quality_df[column_name].plot.box(ax=flat_ax[idx])
```
