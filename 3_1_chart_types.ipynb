```python
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 1000)

test_df = pd.DataFrame({
    "x": x,
    "sin_x": np.sin(x)
})
test_df.sample(5)
```

```python
test_df.plot(figsize=(10.0, 7.5))
```

```python
test_df.plot(x="x", y="sin_x", figsize=(8.0, 5.5))
```

```python
iris_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
iris_df.sample(5)
```

```python
iris_df.describe()
```

```python
iris_df["species"].unique()
```

```python
iris_df.plot.scatter(x="sepal_length", y="sepal_width",
                     figsize=(8.0, 5.5))
```

```python
colors, mapping = pd.factorize(iris_df["species"])
colors, mapping
```

```python
iris_df.plot.scatter(x="sepal_length", y="sepal_width",
                     c=colors, cmap="tab10", colorbar=False,
                     figsize=(8.0, 5.5))
```

```python
iris_df.groupby("species").mean()
```

```python
iris_df.groupby("species") \
    .mean() \
    .plot.barh(cmap="tab10",
               figsize=(8.0, 5.5))
```

```python
iris_df["sepal_length"] \
    .plot.hist(bins=10, figsize=(8.0, 5.5))
```

```python
iris_df["sepal_length"] \
    .plot.density(figsize=(8.0, 5.5))
```

```python
iris_df.plot.hexbin(x="petal_length", y="petal_width", figsize=(12.5, 8.0), gridsize=(13,7))
```

```python
iris_df["species"].value_counts()
```

```python
iris_df["species"].value_counts().plot.pie()
```

```python
iris_df.plot.box()
```

```python
iris_df.groupby("species") \
    .plot.box(vert=False)
```
