```python
import numpy as np
import pandas as pd
```

```python
dataset_df = pd.DataFrame({
    "x": np.linspace(0, 2 * np.pi, 500)
})
dataset_df
```

```python
np.random.seed(2020)

dataset_df = dataset_df \
    .assign(y=2 * dataset_df["x"] + 1) \
    .assign(z=-dataset_df["x"] - 1) \
    .assign(six_x=np.sin(dataset_df["x"])) \
    .assign(random=np.random.normal(scale=3.0, size=500)) \
    .assign(log_x=np.log(dataset_df["x"] + 10e-6))
dataset_df.head(10)
```

```python
dataset_df.plot(x="x", subplots=True, layout=(5,1))
```
