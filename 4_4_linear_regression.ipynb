```python
import pandas as pd
```

```python
xy_df = pd.DataFrame({
    "x": [1, -2, 2, -1, 0],
    "y": [2, -1, 2.5, 0, 1.5],
})
xy_df.plot.scatter(x="x", y="y", s=25, figsize=(10, 7.5))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_line(a, b, x_min=-5, x_max=5):
    x = np.linspace(x_min, x_max, 1000)
    y = a*x + b
    plt.plot(x, y, "-r", c=(.75, .1, .1, .25))
```

```python
xy_df.plot.scatter(x="x", y="y", s=25, figsize=(10, 7.5))
for x in np.linspace(0.75, 1.25, 10):
    draw_line(x, 1.25, -2, 2)
```

```python
wine_quality_df = pd.read_parquet("./data/winequality.parquet")
```

```python
wine_quality_df.head(10)
```

### Trenowanie modelu

```python
wine_quality_df.plot.scatter(x="alcohol", y="quality", figsize=(10.0, 7.5))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
wine_quality_df.reset_index(drop=True, inplace=True)
```

```python
X = wine_quality_df[["alcohol"]]
y = wine_quality_df["quality"]

X.sample(5)
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
lr.coef_, lr.intercept_
```

```python
wine_quality_df.plot.scatter(x="alcohol", y="quality", figsize=(10.0, 7.5))
draw_line(lr.coef_, lr.intercept_, 0, 20)
```
