### Zmniejszenie wymiarowości danych

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
```

```python
credit_cards_2d = tsne.fit_transform(
    credit_cards_df.drop(columns="DEFAULT")
)
credit_cards_2d_df = pd.DataFrame(credit_cards_2d, columns=["tsne_0", "tsne_1"])
credit_cards_2d_df.plot.scatter(x="tsne_0", y="tsne_1", c=credit_cards_df["DEFAULT"], cmap="Wistia", figsize=(12, 7.5))

```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
credit_cards_2d = pca.fit_transform(
    credit_cards_df.drop(columns="DEFAULT")
)
credit_cards_2d_df = pd.DataFrame(credit_cards_2d, columns=["pca_0", "pca_1"])
credit_cards_2d_df.plot.scatter(x="pca_0", y="pca_1", c=credit_cards_df["DEFAULT"], cmap="Wistia", figsize=(12, 7.5))
```
