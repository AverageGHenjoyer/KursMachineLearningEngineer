```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(2020)
x = np.linspace(0, 10, 12)
xy_df = pd.DataFrame({
    "x": x,
    "y": x ** 2 - 7 * x + 2 * np.random.normal(size=12)
})
xy_df
```

```python
xy_df.plot.scatter(x="x", y="y", figsize=(10,7.5))
```

```python
X, y = xy_df[["x"]], xy_df["y"]
```

```python
X_test = np.linspace(-2, 12, 1000).reshape(-1, 1)
```

### ElasticNet

```python
from sklearn.linear_model import ElasticNet
```

```python
en = ElasticNet()
en.fit(X, y)
```

```python
import matplotlib.pyplot as plt
```

```python
xy_df.plot.scatter(x="x", y="y", figsize=(10,7.5))
plt.plot(X_test[:, 0], en.predict(X_test), c="red")
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(X, y)
```

```python
xy_df.plot.scatter(x="x", y="y", figsize=(10,7.5))
plt.plot(X_test[:, 0], dt.predict(X_test), c="red")
```

### Radius Neighbors Regression

```python
from sklearn.neighbors import RadiusNeighborsRegressor
```

```python
rnr = RadiusNeighborsRegressor(radius=2)
rnr.fit(X, y)
```

```python
xy_df.plot.scatter(x="x", y="y", figsize=(10,7.5))
plt.plot(X_test[:, 0], rnr.predict(X_test), c="yellow")
```
