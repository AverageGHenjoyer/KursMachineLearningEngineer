```python
import pandas as pd
```

```python
wine_quality_df = pd.concat([
    pd.read_csv("./data/winequality-red.csv", delimiter=";").assign(color="red"),
    pd.read_csv("./data/winequality-white.csv", delimiter=";").assign(color="white"),
])
```

```python
wine_quality_df.sample(10)
```

```python
wine_quality_df = pd.get_dummies(wine_quality_df)
wine_quality_df.sample(10)
```

```python
wine_quality_df["bound sulfur dioxide"] = \
    wine_quality_df["total sulfur dioxide"] - wine_quality_df["free sulfur dioxide"]
wine_quality_df.sample(10)
```

```python
wine_quality_df.corr().style.background_gradient(cmap="RdYlGn")
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots(4, 4, sharey=False, figsize=(21.0, 32.0))
wine_quality_df.boxplot(by="quality", ax=ax.flatten()[:-2],
                        layout=(4, 4))
```

```python
wine_quality_df.to_parquet("./data/winequality.parquet")
```
