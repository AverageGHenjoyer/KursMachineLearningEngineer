```python
pd.get_dummies(symptoms_df)
```

```python
from sklearn.impute import SimpleImputer
```

```python
mf_imputer = SimpleImputer(strategy="most_frequent")
mf_imputer.fit(symptoms_df)

mf_imputer.transform(symptoms_df)
```

```python
pd.DataFrame(mf_imputer.transform(symptoms_df),
             columns=symptoms_df.columns)
```

```python
const_imputer = SimpleImputer(strategy="constant",
                              fill_value="Unknown")
const_imputer.fit(symptoms_df)
```

```python
pd.DataFrame(const_imputer.transform(symptoms_df),
             columns=symptoms_df.columns)
```
