```python
import pandas as pd
```

```python
wine_quality_df = pd.read_parquet("./data/winequality.parquet")
wine_quality_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(wine_quality_df, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
pipeline = Pipeline( steps=[
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
]
)
```

```python
pipeline.fit(train_df.drop("quality", axis="columns"),
             train_df["quality"])
```

```python
train_pred = pipeline.predict(train_df.drop("quality", axis="columns"))
test_pred = pipeline.predict(test_df.drop("quality", axis="columns"))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(train_df["quality"], train_pred), mean_absolute_error(test_df["quality"], test_pred)
```

```python
mean_squared_error(train_df["quality"], train_pred), mean_squared_error(test_df["quality"], test_pred)
```

```python
pipeline.steps[1][1].coef_, pipeline.steps[1][1].intercept_
```
